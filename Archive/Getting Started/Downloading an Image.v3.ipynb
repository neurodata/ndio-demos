# Downloading an Image
Downloading a single image from a slice of a volume is easy in ndio. Here, we'll download a slice of an image from the Bock11 dataset.

```python
import ndio.remote.neurodata as ND
nd = ND(hostname = 'cloud.neurodata.io')
```

```python
binary_img_data = nd.get_image('bock11', 'image', 11000, 11500, 11000, 11500, z_index=4000, resolution=3)
```

We now have the binary image data stored in the variable `binary_img_data`. We can save it to a file like this:

```python
import ndio.convert.png as ndpng
ndpng.save('bock11-demo.png', binary_img_data)
```

We can also verify that we downloaded the correct image by rendering it in Python. While it's far easier to open this image by navigating to the directory where you saved it (by default, it saves in the same directory as the Python script), we demonstrate showing it in Python below for completeness's sake:

```python
from PIL import Image
Image.open('bock11-demo.png')
```

Note that we could have just as easily used `ndio.convert.tiff` instead of `ndio.convert.png`.
